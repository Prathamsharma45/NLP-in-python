```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])

    return L
```

```python

df['keywords']=df['keywords'].apply(convert)

```

```python
df['genres']=df['genres'].apply(convert)
```

```python
def convert_cast(obj):
    L=[]
    ct=0
    for i in ast.literal_eval(obj):
        if ct!=3:
            L.append(i['name'])
            ct=ct+1
        else:
            break
    return L
```

```python
df['cast']=df['cast'].apply(convert_cast)
```

```python
df['cast'][0]
```

```python
def fetch_director(obj):
    L=[]
    
    for i in ast.literal_eval(obj):
      if i['job']=='Director':
            L.append(i['name'])
    return L
```

```python
df['crew']=df['crew'].apply(fetch_director)
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
ps.stem('walked')
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
df['tags'] = df['tags'].apply(stem)
```
